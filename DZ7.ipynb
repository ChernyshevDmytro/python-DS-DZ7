```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.linear_model import RidgeClassifier
import numpy as np
import pandas as pd

wine_dataset = load_wine()

data = pd.DataFrame(data= np.c_[wine_dataset['data'], wine_dataset['target']], columns= wine_dataset['feature_names'] + ['target'])
print(type(wine_dataset))
data
```

```python
X, y = load_wine(return_X_y=True)

X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=False)

clf = RidgeClassifier().fit(X_train1, y_train1)

clf.score(X_test1, y_test1)
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y, test_size=0.35, random_state=42, shuffle=True)

clf = RidgeClassifier().fit(X_train2, y_train2)

clf.score(X_test2, y_test2)
```

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression

diabetes= load_diabetes()
X4, y4 = load_diabetes(return_X_y=True)
X_train4, X_test4, y_train4, y_test4 = train_test_split(X4, y4, test_size=0.35, random_state=42, shuffle=True)

data2 = pd.DataFrame(data= np.c_[diabetes['data'], diabetes['target']],columns= diabetes['feature_names'] + ['target'])
data2
```

```python
X_train4, X_test4, y_train4, y_test4 = train_test_split(X4, y4, test_size=0.35, random_state=42, shuffle=False)
regressor = LinearRegression().fit(X_train4, y_train4)
regressor.score(X_test4, y_test4)
```

```python
X_train5, X_test5, y_train5, y_test5 = train_test_split(X4, y4, test_size=0.45, random_state=42, shuffle=True)
regressor = LinearRegression().fit(X_train5, y_train5)
regressor.score(X_test5, y_test5)
```

```python
from sklearn.linear_model import SGDRegressor
X_train6, X_test6, y_train6, y_test6 = train_test_split(X4, y4, test_size=0.35, random_state=42, shuffle=True)
regressor = SGDRegressor().fit(X_train5, y_train5)
regressor.score(X_test6, y_test6)
```

```python
best_result = [0.01, 0.1]

t_s = np.arange(0.1, 1, step=0.01)

for i in range(len(t_s)):
  X_train6, X_test6, y_train6, y_test6 = train_test_split(X4, y4, test_size=t_s[i], random_state=42, shuffle=True)
  regressor = LinearRegression().fit(X_train5, y_train5)
  score = regressor.score(X_test5, y_test5)
  #print(t_s[i], score)
if score > best_result[0]:
  best_result[0] = score
  best_result[1] = t_s[i]
best_result
```
